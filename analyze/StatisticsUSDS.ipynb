```python
listK = sd.get_day_data("20190101")
# print listK
```

```python
#第一天涨停，第二天继续上涨
uscMap = {}

#第一天跌停，第二天上涨
dscMap = {}

pre = None
for k in listK:
    if pre==None:
        pre = k
        continue
    if pre.code!=k.code:
        pre = k
        continue
    ds = dscMap.get(k.date)
    us = uscMap.get(k.date)
    if pre.isUS() and None==us:
        us=[k.date,0,0]
        uscMap[k.date]=us
    if pre.isDS() and None==ds:
        ds=[k.date,0,0]
        dscMap[k.date]=ds
    if pre.isUS():
        if k.isUp():
            us[1]=us[1]+1
        else:
            us[2]=us[2]+1
    
    if pre.isDS():
        if k.isUp():
            ds[1]=ds[1]+1 
        else:
            ds[2]=ds[2]+1 

usc_temp = sorted(uscMap.values(), key=lambda v: v[0])
dsc_temp = sorted(dscMap.values(), key=lambda v: v[0])
usc = []
dsc = []
for v in usc_temp:
    usc.append([v[0],v[1]/(v[1]+v[2]+0.0)*100])
for v in dsc_temp:
    dsc.append([v[0],v[1]/(v[1]+v[2]+0.0)*100])
```

```python
params = {}
params["name"] = '行情统计'

params["usc"] = json.dumps(usc)
params["dsc"] = json.dumps(dsc)
```

```python
s = show("StatisticUSDS.html",params)
```
