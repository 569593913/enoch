```python
from Selector import *
import sys
import time
from view.HTML import show
import json
from indicators.MA import *
from simulation.Simulator import *
sys.path.append("../data")
import KDataDao as kds
```

```python
def getTimeInt(date):
    return int(time.strftime("%s",time.strptime(date, '%Y-%m-%d'))) * 1000
def firgueF(code):
    """显示股票"""
    #获取数据
    listK = kd.get_k_data(code,"2014-01-01")
    data = []
    volume = []
    i = 0;
    line = []#[[getTimeInt(listK[0].date),listK[0].low],[getTimeInt(listK[len(listK)-1].date),listK[len(listK)-1].low]]
    v = None
    preClose = listK[0].close
    preI = 0;
    nextI = 0;
    ma10 = MA(30)
    line10 = []
    ma60 = MA(60)
    line60 = []
    earningsLine = []
    sim = Simulator()
    for k in listK:
        d = getTimeInt(k.date)
        
        #模拟交易
        if sim.isHold(k.code) and k.isDown():
            sim.sell(k.code,d,k.close,)
        if k.isUp:
            sim.buy(k.code,d,k.close)
        sim.fresh(code,k.close)
        earningsLine.append([d,sim.earnings()])
        
        #K线
        data.append([ d,k.open,k.high,k.low,k.close])
        preClose = k.close
        #画线
        preI = i-1;
        if preI<0:
            preI = 0
        nextI = i+1
        if nextI>=len(listK):
            nextI = len(listK)-1
        if listK[preI].low>k.low and listK[nextI].low>k.low :
            if v!=None and v!=k.low:
                line[-1] = None
            v = k.low
        if v!=None and (v>k.low or (k.low-v)/v > 0.35):
            v = None
        line.append([d,v])
        #成交量
        volume.append([d,k.volume])
        #均线
        ma10.add(k.close)
        ma60.add(k.close)
        line10.append([d,ma10.last()])
        line60.append([d,ma60.last()])
        i+=1
        
    #获取开始显示的时间
    mDate = 0
    if len(data)>0:
        if len(listK)/2>-250:
            mDate = data[-250][0]
        else:
            mDate = data[len(listK)/2][0]
    
    #显示
    params = {}
    params["line10"] = json.dumps(line10)
    params["line60"] = json.dumps(line60)
    params["line"] = json.dumps(line)
    params["code"] = code
    params["mDate"] = mDate
    params["data"] = json.dumps(data)
    params["volume"] = json.dumps(volume)
    params["earningsLine"] = json.dumps(earningsLine)
    clear_output()
    show("stock.html",params)
```

```python
class KBlance(Selector):
    """红绿红"""
    
    def analyze(self,list):
        """分析股票"""
        if True:
            return True
        if (
                        abs((list[-1].maxOC() - list[-3].maxOC()) / list[-3].maxOC()) < 0.03
                and list[-1].ratio() > 0.01
            and list[-2].ratio() < -0.01
        and list[-3].ratio() > 0.01):
            return True
        return False
    
    def firgue(self,code):
        """显示股票"""
        firgueF(code)
```
