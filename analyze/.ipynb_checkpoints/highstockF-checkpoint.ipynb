```python
import  os
import sys
sys.path.append("../data")
import KDataDao as kd
import charts
import time
from IPython.display import display,clear_output
import ipywidgets as widgets
```

```python
#从数据库获取数据
def getCandlestickData(code):
    listK = kd.get_k_data(code,"2017-01-01")
    data = []
    volume = []
    line = []#[[listK[len(listK)-10].date,listK[len(listK)-10].open],[listK[len(listK)-1].date,listK[len(listK)-1].open]]
    i = 0;
    for k in listK:
        d = int(time.strftime("%s",time.strptime(k.date, '%Y-%m-%d'))) * 1000
        data.append([ d,k.open,k.high,k.low,k.close])
        if i%4==0:
            line.append([d,k.open])
        else:
             line.append([d,0])
        i+=1
        volume.append([d,k.volume])
    return data,volume,line
```

```python
#highstock数据,参数封装
def getHighstockData(code):
    data,volume,line = getCandlestickData(code)
    #获取中间数据的时间
    mDate = 0
    if len(data)>0:
        mDate = data[len(data)/2][0]
    options = {
            "plotOptions": {
                 "candlestick": {
                            "color": '#ffffff',
                            "lineColor": '#7CB5EC',
                            "upColor":'red',
                            "upLineColor": 'red',
                        }
            },
            "chart": {
                "height": 500,
            },
            "rangeSelector": {
               "enabled":False
            },
            "tooltip": {
                "formatter": "@#function () {\
                    return '<b>' + Highcharts.dateFormat('%Y-%m-%d',this.x) + '</b>'\
                    +'<br>'+'<b>'+'code: '+'</b>' + this.points[0].series.name\
                    +'<br>'+'<b>'+'open: '+'</b>' + this.points[0].point.open\
                    +'<br>'+'<b>'+'close: '+'</b>' + this.points[0].point.close\
                    +'<br>'+'<b>'+'hight: '+'</b>' + this.points[0].point.high\
                    +'<br>'+'<b>'+'low: '+'</b>' + this.points[0].point.low\
                    +'<br>'+'<b>'+'ratio: '+'</b>' + (this.points[0].point.open==0?'-':\
                    ((this.points[0].point.close-this.points[0].point.open)/this.points[0].point.open*100).toFixed(2))+'%'\
                    +'<br>'+'<b>'+'volume: '+'</b>' + this.points[0].point.y\
                }@#",
                "xDateFormat":'%Y-%m-%d ',
                "shared": True
            },
            "title": {
                "text": code+' Historical'
            },
            "xAxis": {
                "type": 'datetime',
                "min": mDate
              },
            "yAxis": [{
                "height": '60%',
                "lineWidth": 2
            }, {
                "top": '65%',
                "height": '35%',
                "offset": 0,
                "lineWidth": 2
            }]

        }
    series=[
            {
                "type": 'candlestick',
                "name": code,
                "data": data,
                "plotOptions": {
                     "candlestick": {
                                "color": '#ffffff',
                                "lineColor": '#7CB5EC',
                                "upColor":'red',
                                "upLineColor": 'red',
                            }
                },
            }, {
                "type": 'column',
                "name": 'Volume',
                "data": volume,
                "yAxis": 1
            }
    ]
    return options,series
```

```python
#画K线图
def firgue(code):
    opt,ser = getHighstockData(code)
    clear_output()
    display(charts.plot(ser,opt,show="inline",stock=True,height=500))
```

```python
#按钮
nextBtn = widgets.Button(description="next")
nextPre = widgets.Button(description="pre")
text = widgets.HTML(
    value="selecting....",
    placeholder='select',
    description='select',
)
def next_clicked(a):
    Stock.figureNext()
    text.value = str(Stock.idx+1)+"/"+str(len(Stock.codes))
def pre_clicked(a):
    Stock.figurePre()
    text.value = str(Stock.idx+1)+"/"+str(len(Stock.codes))
display(widgets.Box([nextPre,nextBtn,text]))
nextBtn.on_click(next_clicked)
nextPre.on_click(pre_clicked)


text.value = str(Stock.idx)+"/"+str(len(Stock.codes))

```
