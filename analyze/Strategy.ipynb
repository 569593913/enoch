```python
class AStrategy:
    """
    K线策略
    """
    def __init__(self):
       self.point = 0;     
    def decide(self,list,trader,i):
        """
        策略
        :param list: 测试数据
        :param trader: 交易者
        :param i: 当前在list中的位置
        :return:
        """
        k = list[i]
        #卖策略
        if  (
            trader.isHold(k.code) 
            and 
            ( 
                self.point > k.maxOC() 
                or ( k.isDown() and list[i-1].isDown() )
                or ( k.isDown() and k.close < list[i-2].low)
            )
            ):
#             print "date:%s point:%s maxOC:%s k:%s l:%s" \
#             % (datetime.utcfromtimestamp(k.date/1000),self.point , k.maxOC(),k.isDown() , list[i-1].isDown())
            trader.sell(k.code,k.date,k.close)
        #买策略
        if (
            abs((k.maxOC() - list[i-2].maxOC()) / list[i-2].maxOC()) < 0.02
            and abs((k.maxOC() - list[i-1].maxOC()) / list[i-1].maxOC()) < 0.02
            and k.isUp()
            and (list[i-1].isUp() or list[i-2].isUp())
            and abs(k.ratioAbs() - list[i-2].ratioAbs()) < 0.015
            and abs(k.ratioAbs() - list[i-1].ratioAbs()) < 0.015
            and k.ratioAbs() < 0.05
            ):
#             print "%s,r:%s,r1:%s,r2:%s" %(k,k.ratio(),list[i-1].ratio(),list[i-2].ratio())
            self.point = k.minOC()
            trader.buy(k.code,k.date,k.close)
```
